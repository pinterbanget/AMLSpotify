```python
data = pd.read_csv('Spotify_Dataset_V3.csv', delimiter=';')
data.head()
```

#### Top global 2017-2023

kalo artist poinnya sama semua regardless of colabs

```python
artist_score = data[["Artist (Ind.)", "Points (Total)"]]
most_charting_artists = artist_score.groupby("Artist (Ind.)").sum()
most_charting_artists = most_charting_artists.sort_values("Points (Total)", ascending=False)
most_charting_artists[:100]
```

kalo collab berdua poinnya bagi 2, bertiga poinnya bagi 3, ber-n poinnya bagi n

```python
artist_score = data[["Artist (Ind.)", "Points (Ind for each Artist/Nat)"]]
most_charting_artists = artist_score.groupby("Artist (Ind.)").sum()
most_charting_artists = most_charting_artists.sort_values("Points (Ind for each Artist/Nat)", ascending=False)
most_charting_artists[:100]
```
